## Chapter 1: The machine learning landscape


#### Supervised vs Unsupervised learning:

Supervised learning: the training data are already "labeled" with the solutions. Classification and regression (for prediction) are examples of tasks. The most important algorithms for supervised learning are:
- k-Nearest Neighbors
- Linear Regression
- Logistic Regression
- Support Vector Machines (SVMs)
- Decision Trees and Random Forests
- Neural networks

Unsupervised learning: the training data are unlabeled, the algorithm learns without a "supervisor". Examples of tasks are clustering, anomaly detection, associate rule learning, amongst many others. Some important unsupervised learning algorithms are:
- Clustering
 - K-Means
 - DBSCAN
 - Hierarchical Cluster Analysis (HCA)
- Anomaly detection and novelty detection 
 - One-class SVM
 - Isolation Forest
- Visualization and dimensionality reduction
 - Principal Component Analysis (PCA)
 - Kernel PCA
 - Locally-Linear Embedding (LLE)
 - t-distributed Stochastic Neighbor Embedding (t-SNE)
- Association rule learning
 - Apriori
 - Eclat
 
Semisupervised learning: the training data are a mix of (mostly) unlabeled data and (some) labeled data.

Reinforcement learning: the learning system (agent) is rewarded or penalized for its actions and optimizes a strategy based on this feedback.

#### Batch vs online learning

Batch learning: system is incapable of learning incrementally, must be trained on all the data at once, hence "offline", before being launched into production. Good if system doesn't need to be updated "on the fly". Even once a day or week updates are ok. Each update/learning instance requires all the data, so it must all be kept.

Online learning: system is fed data sequentially, keeping each learning step fast and cheap. Good when data is received continuously and system needs to change rapidly or autonomously. Data can also be discarded once system has trained on it. Learning rate affects the memory of the system, and the sensitivity to new data.

#### Instance-based vs model-based learning

Instance-based learning: new instances are simply compared to old instances the system was trained on (and "memorized").

Model-based learning: a model is built from training data, which can then be used to make predictions from new instances. 

#### Testing and validating

Common to train on 80% of the data (training set) and to hold 20% of the data (test data) to evaluate the model's performance on after training. The error is called generalization error, or out-of-sample error. Exact break-down between training and test data depends on the size of the data set.

Also, hold-out and cross validation techniques allow you to use a subset of the training data to select a model and another set to train it.
